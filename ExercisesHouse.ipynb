### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Create 3 differents Series, each of length 100, as follows: 
1. The first a random number from 1 to 4 
2. The second a random number from 1 to 3
3. The third a random number from 10,000 to 30,000

```python
df1 = pd.DataFrame(np.random.randint(1,4,size=(100, 1)), columns=list('a'))
df2 = pd.DataFrame(np.random.randint(1,3,size=(100, 1)), columns=list('b'))
df3 = pd.DataFrame(np.random.randint(10000,30000,size=(100, 1)), columns=list('c'))

```

### Step 3. Let's create a DataFrame by joinning the Series by column

```python
frames = [df1, df2, df3]

df = pd.concat(frames, axis=1, sort=False)

df
```

### Step 4. Change the name of the columns to bedrs, bathrs, price_sqr_meter

```python
df.rename(columns={"a": "bedrs", "b": "bathrs", "c":"price_sqr_meter"})
```
